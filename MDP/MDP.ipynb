```python
# pip install gym
# pip install gymnasium 
help("gym")

import gymnasium as gym
import numpy as np
```

```python
env = gym.make("FrozenLake-v1",  
               render_mode = "ansi", 
               is_slippery = False) 
    # charcater-mode
    # deterministic environment
print(env.observation_space) # 16 states (상태공간)
print(env.action_space) # 4 actions (행동공간)
```

```python
state, info = env.reset() # 초기 상태
print(env.render()) # 초기 상태 출력
```

```python
while True: 
    action = env.action_space.sample() # 행동 선택 (랜덤)
    state, reward, terminated, truncated, info = env.step(action) # 행동 수행
    print(env.render()) # 상태 출력
    if terminated or truncated: 
        break
    # 난수를 사용하기 때문에 실행할 때마다 다른 결과를 얻음
```

```python
env.close()
```
