```python
# pip install gym
# pip install gymnasium 
# help("gym")

import gymnasium as gym
import numpy as np
```

### 01. FrozenLake 과업의 MDP 사이클 확인

```python
# FrozenLake 과업의 MDP 사이클 확인
env = gym.make("FrozenLake-v1",  
               render_mode = "ansi", 
               is_slippery = False) 
    # charcater-mode : ansi text mode(american national standard code for information interchange)
    # is_slippery : deterministic environment
print(env.observation_space) # 16 states (상태공간)
print(env.action_space) # 4 actions (행동공간)
```

```python
n = 100000 # 에피소드 개수
env = gym.wrappers.TimeLimit(
    env, max_episode_steps = 100
)
```

```python
state, info = env.reset() # 초기 상태
print(env.render()) # 초기 상태 출력
```

### 02. 랜덤 정책의 기대이익을 통해 비효율 확인

```python
# action을 랜덤하게 선택하여 MDP 사이클 확인하여 
# 랜덤 정책과 최적 정책의 기대이익(expected return) 비교

episodes = []
NUM_EPISODES = 100  # 원하는 에피소드 수

for _ in range(NUM_EPISODES):
    episode = []  # 에피소드 기록
    state, info = env.reset()  # 초기 상태
    episode.append([None, state])  # 초기 상태 기록

    while True:  # 한 에피소드 내 스텝 반복
        action = np.random.choice(
            [0, 1, 2, 3], 
            1, 
            p=[0.25, 0.25, 0.25, 0.25]
        )[0]  # 행동 선택 (랜덤)
        
        state_01, reward, terminated, truncated, info = env.step(action)  # 행동 수행
        state = state_01  # 상태 업데이트

        episode.append([action, reward, state])
        
        if terminated or truncated:  # 에피소드 종료 조건
            break  # 에피소드 종료 → 다음 에피소드로
        
        # 난수를 사용하기 때문에 실행할 때마다 다른 결과를 얻음

    episodes.append(episode)  # 에피소드 리스트에 원소 추가
```

```python
# 랜덤 정책의 기대이익을 통해 비효율 확인
expected_return = sum([e[-1][1] for e in episodes]) / n
print(f"랜덤 정책의 기대이익 : {expected_return:.8f}")
```
